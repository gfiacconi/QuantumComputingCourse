The **Block Sphere** is a *vector space* constituted by the orthonormal base $|0\rangle\ |1\rangle\ $ with which we can construct the Superposition of the states by simply making a linear combination of them.
The generic Qubit is rappresented by:

$$|\psi\rangle =\alpha\|0\rangle+\beta|1\rangle\$$
with $\vert\alpha\vert^2+\vert\beta\vert^2 = 0$

Each *operation* with the Qubits corrispond to a ***Projection*** of $|\psi\rangle$ on the orthonormal base $|0\rangle\ |1\rangle\ $ and it **will not produce a be deterministic result** but a ***probabilistic*** result.

>In fact $\vert\alpha\vert^2+\vert\beta\vert^2 = 0$ has a probabilistic meaning.
>- $\vert\alpha\vert^2$ is the probability that $|\psi\rangle$ decay on $|0\rangle$
>- $\vert\beta\vert^2$ is the probability that $|\psi\rangle$ decay on $|1\rangle$

To describe a particular state of a particle we can introduce a complex vector space and associate at each state of the motion of the particle a vector in that space. This procedure was thought by professor *Werner Heisenberg* in his ***Matricial Formulation of the QUantum Mechanic in 1928**. 
So now in the complex vector space we can us all the properties of the complex vectors and we call it **H**.

**H** now has all the significant wave functions and the identity function rapressented all by vectors.
Those vectors are linearly indipendent.
In general the Complex vector space can have two types of dimension
- finite, and it correspond to the number of L.I. vectors
- infinite
in our case, **H** has all significant solutions of De Broglie wave functions.

In quantum computing is common the BraKet notation or also called Dirac Notation
1. **Bra** 
    >- $\langle\ 0 \vert =\begin{bmatrix} 1 & 0 \end{bmatrix}$

    >- $\langle\ 1 \vert = \begin{bmatrix} 0 & 1 \end{bmatrix}$
2. **Ket** 
    >- $|0\rangle =\begin{bmatrix} 1 \\ 0 \end{bmatrix}$

   >- $|1\rangle = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$
___
Now let's introduce the operations.
#### 1. Inner Product
- the main result is to give a Scalar to a couple of vector
let's see how it works:\
\
$\langle\ 0 \vert\ 1 \rangle = \begin{bmatrix} 1 & 0 \end{bmatrix} \begin{bmatrix} 1 \\ 0 \end{bmatrix} = (1 * 0)(0 * 1) = 0 $
___
#### 2. Outer Product
$|0\rangle\langle\ 1 \vert= |0\rangle\otimes\langle\ 1 \vert  = \begin{bmatrix} 1 \\ 0 \end{bmatrix} \otimes \begin{bmatrix} 0 & 1 \end{bmatrix} = \begin{bmatrix} 0 & 1 \\ 0 & 0 \end{bmatrix}$
___
#### 3. Norms
- is the distance from the origin
$ \vert\vert\vert\ v\rangle \vert\vert = \sqrt{\langle\ v \vert\ v \rangle} $
___
#### 4. Tensor product
- it helps us to combine vector spaces in bigger vector spaces
- it describes the interaction between 2quantum system\
\
$ |0\rangle\otimes\vert\ 1\rangle = |0\rangle\vert\ 1\rangle = |0 1\rangle = \begin{bmatrix} 1 \\ 0 \end{bmatrix}\begin{bmatrix} 0 \\ 1 \end{bmatrix} = \begin{bmatrix} 1*0 \\ 0*0 \end{bmatrix}= \begin{bmatrix} 0 \\ 0 \end{bmatrix} $

___

We will use all of that properties to operate with quantum computers.

## Quantum Operators üßÆ  and Quantum Logical Ports üö™üß†
### Hadamard Port ü™ô
This port produce what is called **Superposition** state of the different state that we have in our register.

**Superposition** means that we have the same **probability** for all of the states that we are watching, and that's why this port is called *Coin-Toss port*.

Let's have an example:

```python
from qiskit import QuantumCircuit, assemble, Aer #library to create quantum circuits
from math import pi, sqrt #usueful math tools
from qiskit.visualization import plot_bloch_multivector, plot_histogram #library to plor
sim = Aer.get_backend('aer_simulator')

qc1= QuantumCircuit(1) #we built a Single Qubit Gates set at |0> state

qc1.draw() #we draw it 
```

```python
# Let's see |0> state where we set that we start:
qc1.save_statevector()
qobj = assemble(qc1)
state = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(state)
```

```python
#we applied an H-gate on our |0> qubit
qc2= QuantumCircuit(1)

qc2.h(0) 
qc2.draw()
```

```python
# Let's see the result of an H-gate on our |0> qubit:
qc2.save_statevector()
qobj = assemble(qc2)
state = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(state,title="H-gate on our |0> qubit")
```

We can notice that our vector set up at the $\vert 0\rangle$ state has moved...but why???ü§®

By appying the Hadamard port we rotated the vector.

In fact the Hadamard port is an operator which is a function over a space of physical states onto another space of physical states...but we can say that is just a fuction that moves our vector in our space ü•≤.

If we do the math in fact we can see why we have this result.

Let's start the dimostration by showing the **Hadamard-operator**:
>$$ H = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix} $$ 

Which we can decompose in:
> $$\frac{\vert\ 0\rangle+ i\vert\ 1\rangle}{\sqrt{2}} \langle 0\vert + \frac{\vert\ 0\rangle- i\vert\ 1\rangle}{\sqrt{2}} \langle 1\vert$$


Now we apply it to the $\color{red}{\vert 0\rangle}$ that we set as our initial state:
>$$ H\color{red}{\vert 0\rangle} = \frac{\vert\ 0\rangle+ i\vert\ 1\rangle}{\sqrt{2}} \langle 0\vert \color{red}{0 \rangle} + \frac{\vert\ 0\rangle- i\vert\ 1\rangle}{\sqrt{2}} \langle 1\vert\color{red}{ 0 \rangle} $$  

we can see that:
$$\langle 0\vert0 \rangle=\begin{bmatrix} 1 & 0 \end{bmatrix}\begin{bmatrix} 1 \\ 0 \end{bmatrix} = (1*1) + (1*0) = 1$$
and
$$\langle 0\vert1 \rangle=\begin{bmatrix} 1 & 0 \end{bmatrix}\begin{bmatrix} 0 \\ 1 \end{bmatrix} = (1*0) + (0*1) = 0$$

So we are left with just:
>$$ H\vert 0\rangle = \frac{\vert\ 0\rangle+ i\vert\ 1\rangle}{\sqrt{2}} \langle 0\vert 0 \rangle $$

which is just the  $\vert +\rangle$ state

In fact if we bring it back we can see that is just the same:

```python
#plotting the ket + state
oneState = [1,0,0] # [X, Y, Z]
plot_bloch_vector(oneState,title='|+> State',figsize=(4,4)) # Bloch Vector with spherical coordinates
```

so the **Hadamard Gate** produce a rotaion of $\frac{\pi}{2}$ along the Y-axis if applied to the initial state set up at $\vert\ 0\rangle$.

But what will append if we applie it to an initial state set up at $\vert\ 1\rangle$???üßê

Let's begin with just making our circuit.

```python
qc3= QuantumCircuit(1) #we built a Single Qubit Gates set at |1> state
qc3.x(0) #i have applied an X-Pauli-Gate just to put it on |1>, we will see later what X-gate is...
qc3.draw() #we draw it 
```

```python
# Let's see the result of an H-gate on our |0> qubit:
qc3.save_statevector()
qobj = assemble(qc3)
state = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(state,title="H-gate on our |1> qubit")
```

```python
qc4= QuantumCircuit(1)
qc4.x(0)
#we applied the Hademard-Gate to the |1> initial state
qc4.h(0) 
#we start plotting
qc4.save_statevector()
qobj = assemble(qc4) 
state = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(state,title="H-gate on our |1> qubit")
```

As you can see is the opposite of the $\vert 0\rangle$ so it corrispond to the $\vert -\rangle$ state.

So the **Hadamard Gate** produce a rotaion of  $\frac{\pi}{2}$ along the X-axis if applied to the initial state set up at $\vert\ 1\rangle$.
